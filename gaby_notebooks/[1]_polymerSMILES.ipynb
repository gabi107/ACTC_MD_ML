# Get Morgan Fingerprint 

### XLAPn4_45p

```python
fpA = AllChem.GetMorganFingerprintAsBitVect(XLAPn4_45p, radius=2, nBits=2048, useChirality=True)
#fpA = AllChem.GetMorganFingerprintAsBitVect(XLAPn4_45p, radius=2, nBits=2048, bitInfo=bi)
fpA
```

```python
bi ={}
fpA = AllChem.GetMorganFingerprintAsBitVect(XLAPn4_45p, radius=2, nBits=2048, bitInfo=bi)
fp_arr = np.zeros((1,))
DataStructs.ConvertToNumpyArray(fpA, fp_arr)
np.nonzero(fp_arr)
#on_bits = list(fpA.GetOnBits())
```

```python
prints = [(XLAPn4_45p, x, bi) for x in fpA.GetOnBits()]
prints

Draw.DrawMorganBits(prints, molsPerRow=7, legends=[str(x) for x in fpA.GetOnBits()])
```
